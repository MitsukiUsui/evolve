```python
from Bio.Blast import NCBIXML
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#filepath="../out/test_genus_872.xml"
#filepath="/data/mitsuki/out/altorf/evolve/result/GCF_000022125.1_ASM2212v1.xml"
filepath="/data/mitsuki/out/altorf/evolve/result/GCF_000022125.1_ASM2212v1.filtered.xml"

dct_lst=[]

with open(filepath) as f:
    blastRecords = NCBIXML.parse(f)
    print("DONE loading")
    plt.figure(figsize=(8,8))

    for rec in blastRecords:
        queryName=rec.query
        category=queryName.split('|')[0]
        queryLength=rec.query_length
        for alignment in rec.alignments:
            sbjctName=alignment.title
            sbjctLength=alignment.length
            for hsp in alignment.hsps:
                if hsp.expect<1e-5:
                    dct={}
                    xs=hsp.query_start/queryLength
                    ys=hsp.sbjct_start/sbjctLength
                    xe=hsp.query_end/queryLength
                    ye=hsp.sbjct_end/sbjctLength
                    
                    dct["expect"]=hsp.expect
                    dct["query_name"]=queryName
                    dct["sbjct_name"]=sbjctName
                    dct["query_length"]=queryLength
                    dct["sbjct_length"]=sbjctLength
                    dct["query_start"]=hsp.query_start
                    dct["quer_end"]=hsp.query_end
                    dct["sbjct_start"]=hsp.sbjct_start
                    dct["sbjct_end"]=hsp.sbjct_end
                    
                    dct["category"]=category
                    
                    dct["xs"]=xs
                    dct["ys"]=ys
                    dct["xe"]=xe
                    dct["ye"]=ye
                    dct_lst.append(dct)
                    plt.plot((xs,xe),(ys,ye),'#639df9')
    plt.show()
```

```python
df=pd.DataFrame(dct_lst)
df=df[["category","query_length","sbjct_length","xs","xe","ys","ye","query_name","sbjct_name"]]
df
```
